```r
# Alternative: Building for loops in a function to impute the means for NA values
impute_means = function(data,columns){
                                        for(i in 1: length(columns)){
                                        #print(columns[i])
                                        data[ ,columns[i]][is.na(data[, columns[i]])] = mean(data[ , columns[i]], na.rm = TRUE)
                                        }
                                        return(data) # return the cleaned data
                                    }
```

# 4. Logistic Regression

```r
library("broom")
```

```r
titanic_logreg = titanic2
```

```r
str(titanic_logreg)
```

```r
titanic_logreg$Embarked[is.na(titanic_logreg$Embarked)]

# Create new dataframe that remove NA for Embarked
titanic_logreg1 = titanic_logreg[titanic_logreg$Embarked %in% c('C','Q','S') , ]

summary(titanic_logreg1)
```

```r
# Check that there is no NA values for dataframe 'titanic_logreg1'
colMeans(is.na(titanic_logreg1))
```

## 4.1 Train-Test split

```r
set.seed(123456)
train_ind = sample(1: nrow(titanic_logreg1), size = 0.7 * nrow(titanic_logreg1))
```

```r
train_set = titanic_logreg1[train_ind, ]
test_set  = titanic_logreg1[-train_ind, ]

nrow(titanic_logreg1)
nrow(train_set)
nrow(test_set)
```

## 4.2 Build 1st Model using using Logistics Regression

```r
log_model1 = glm(formula = Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare + Embarked, 
                 data = train_set, family = 'binomial')

summary(log_model1)
```

Q) If the coefficients of the estimates significant to be used for the above regression model?

From the output of the model using logistic regression for the selected variables, 

As the p-value for 4 variables namely 'Parch', 'Fare', 'EmbarkedQ' and 'EmbarkedS' are more than 0.05, the coefficients of the estimates from these variables are not significant to be used for the regression model. 

Pclass2, Pclass3, Sexmale, Age and SibSp coefficient estimate are significant as the p-value is less than 0.05

## 4.3 Build second model using Logistics Regression

```r
log_model2 = glm(formula = Survived ~ Pclass + Sex + Age + SibSp, 
                 data = train_set, family = 'binomial')

summary(log_model2)
```

Q) If the coefficients of the estimates significant to be used for the 2nd logistics regression model?

From the output of the 2nd model using logistic regression for the selected variables, 

Pclass2, Pclass3, Sexmale, Age and SibSp coefficient estimate are significant as the p-value is less than 0.05

```r
head(titanic_logreg1)
```

## 4.3 Predict for the probability of survival for a male aged 35 under Pclass 1 with 1 spouse

```r
predict(object = log_model2, 
        newdata = data.frame(Pclass = '1', Sex = 'male', Age = 35, SibSp = 1), 
        type = 'response')
```

## 4.4 Model Evaluation

```r
log_model2 %>%
    augment()
```

```r
model_result = log_model2 %>%
                    augment() %>%
                    select(c(2:7)) %>%
                    cbind(prob = predict(log_model2, type='response')) %>% #create prob column
                    mutate(odds_ratio = prob/(1-prob)) %>%
                    mutate(log_odds = log(odds_ratio)) %>%
                    mutate(pred = ifelse(prob >= 0.5, 'Yes', 'No'))

model_result

# Column 1 to 6 : upgraded, purchases, extraCards
# Column 7 to 13: .fitted, .resid, .std.resid, .resid, .std.resid, .hat, .sigma, .cooksd
```

## 4.6 Create a function to change binary to 'No' and 'Yes'

```r
NoYes = function(df, col){
        for(i in 1: length(col))
                {
                    df[ ,col[i]] = ifelse(df[ ,col[i]] == '0', 'No', 'Yes')
                }
            return(df)
        }
```

```r
model_result_v1 = model_result %>%
                    mutate(survived_result = model_result$Survived)

head(model_result_v1,10)

class(model_result_v1$survived_result)
```

Convert new column 'survived_results' from numbers to NoYes

```r
model_result_v2 = NoYes(df = model_result_v1, col = c('survived_result'))

head(model_result_v2, 10)
```

## 4.7 Confusion Matrix

Build a confusion matrix to compare the survived_result and predicted result from model built by logistics regression

```r
confusion_matrix = table(model_result_v2$survived_result, model_result_v2$pred)
prop.table(confusion_matrix)

#Adding names to the dimension for easy reference
names(dimnames(confusion_matrix)) = c('survived_result','pred')
confusion_matrix 
```

## 4.7.1 Accuracy & Recall of base model based on train_set

```r
accuracy_train_glm1 = sum(diag(confusion_matrix)) / sum(confusion_matrix)
accuracy_train_glm1
```

```r
recall_train_glm1 = confusion_matrix[2,2] / (confusion_matrix[2,1] + confusion_matrix[2,2])
recall_train_glm1
```

The model accuracy of the train_set1 is calculated to be 0.769
The Recall of the train_set1 is calculated to be 0.701

## 4.7.2 Area Under Curve (AUC) for train_set

```r
#install.packages("pROC")
library(pROC)
```

```r
roc(factor(model_result_v2$survived_result, ordered = TRUE), factor(model_result_v2$pred, ordered = TRUE))
```

Using roc function, Area under the curve is 0.7726 for 2nd model based on the train set.

AUC for this base model based on train set is acceptable as it is greater than 0.7
